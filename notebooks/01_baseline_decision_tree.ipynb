
# Template de Exemplo: Treinamento de Modelo utilizando o Pipeline Centralizado
# Este notebook serve como guia para os integrantes do grupo iniciarem o desenvolvimento de seus respectivos modelos.

```python
# 1. Importando o pipeline e funções de métricas do Sklearn
import matplotlib.pyplot as plt 
from pipeline import run_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
# 2. Carregando os dados perfeitamente tratados e padronizados
X_train, X_test, y_train, y_test = run_pipeline()

print("Dados importados com sucesso!")
print(f"Formato do Treino: {X_train.shape}")
print(f"Formato do Teste: {X_test.shape}")
```

```python
# 3. Treinando um Modelo de Exemplo (Árvore de Decisão)
# Nota: Usamos o random_state fixado para garantir reprodutibilidade
dt_baseline = DecisionTreeClassifier(random_state=42)
dt_baseline.fit(X_train, y_train)

print("Modelo treinado com sucesso!")
```

```python
# 4. Fazendo as previsões no conjunto de Teste
y_pred = dt_baseline.predict(X_test)
```

```python
# 5. Avaliando os Resultados (Foco no Recall)
print("--- Relatório de Classificação ---")
print(classification_report(y_test, y_pred))
```

```python
# 6. Gerando a Matriz de Confusão
print("--- Matriz de Confusão ---")
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Benigno", "Maligno"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de confusão - Decision Tree")
plt.show()
```
